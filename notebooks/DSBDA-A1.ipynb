# 5. Data Formatting and Data Normalization

```python
# Handling missing values

# 1. fillna() : method which we can use replace null(NaN) values with some other data like 0 (applies for all Null values), 
# or we can perform forward fill(ffill), backward fill(bfill).

# 2. interpolate() : performs linear interpolation(by default but you can change it as per need) 
# and replaces the null with the result.

# 3. dropna() : It will drop all the rows that have NaN values. but using this other data also get lost.

# 4. replace() : Replace all given values with specified values
```

```python
df["Cabin"] = df["Cabin"].replace(to_replace=np.nan, value="unknown")
df
```

```python
df["Age"] = df["Age"].interpolate()
df
```

```python
df["Embarked"].fillna(method="ffill",inplace=True)
df
```

```python
# df.dropna() # drop all rows with at least one NaN
# df.dropna(axis = 1) # drop all columns with at least one NaN
# df.dropna(how = 'all') # drop all rows with all NaN values
```

```python
# Changing Datatypes
df["Age"] = df["Age"].astype('int64')
df.dtypes
```
